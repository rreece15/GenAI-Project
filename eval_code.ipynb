```python
from datasets import load_from_disk
from tqdm import tqdm
from transformers import AutoModelForCausalLM, AutoTokenizer, logging
from peft import PeftModel
from datasets import load_from_disk
import eval.code as eval_code
import torch
```

```python
model_id = "mistralai/Mistral-7B-v0.1"

model = AutoModelForCausalLM.from_pretrained(model_id,
                                             torch_dtype=torch.float16,
                                             device_map="auto")
tokenizer = AutoTokenizer.from_pretrained(model_id)

model = PeftModel.from_pretrained(model, "predibase/gsm8k")
```

```python
from datasets import load_dataset

ds = load_dataset("openai/openai_humaneval")
```

```python
ds
```

```python
batch_size = 4
k = 10
per_sample = 5
max_new_tokens = 256
temperature = 0.2
candidate_file = "data/generated_results/gsm8k.json"
tokenizer = AutoTokenizer.from_pretrained(model_id)
dataset = ds["test"]
logging.set_verbosity_error()
```

```python
eval_code.generate_candidates(
    model=model,
    tokenizer=tokenizer,
    dataset=dataset,
    batch_size=batch_size,
    per_sample=5,
    k=k,
    max_new_tokens=max_new_tokens,
    temperature=temperature,
    save_path=candidate_file,
)
```

```python
pass_k = [1,5]
candidates = eval_code.load_candidates(candidate_file)
eval_code.evaluate_candidates(candidates, k=pass_k)
```
